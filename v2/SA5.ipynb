```python
# See: https://stackoverflow.com/questions/53435352/animate-matplotlib-parametric-example
import matplotlib
matplotlib.use('Agg')
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import animation, rc
import matplotlib.pyplot as plt
import math
from celluloid import Camera

import functionsV2
import v2
```

```python
animTStar = functionsV2.animate(v2.x, v2.y, v2.tStarArray, "t*(s)")
```

```python
plt.rcParams.update({'figure.max_open_warning': 0})
figTStar = plt.figure()
figTStar.suptitle('$t*(s)$', fontsize=12)
axTStar = figTStar.gca()
axTStar.set_xticks(np.arange(-0.5,2.5,0.5))
axTStar.set_yticks(np.arange(-1.5,2.0,0.5))
axTStar.set(xlabel='x(t)', ylabel='y(t)',title='header')
axTStar.grid()
axTStar.axhline(y=0, color='k')
axTStar.axvline(x=0, color='k')
cameraTStar = Camera(figTStar)
x=[]
y=[]
r=[]
s=[]

# Prepare arrays r, s
for t in np.arange(0, 1.01, 0.01):
    r += [0.5 + t*(0.3 + 3.9*t - 4.7*t**2)]
    s += [1.5 + t*(0.3 + 0.9*t - 2.7*t**2)]

# Prepare arrays x, y and animate
for t in v2.tStarArray:
    x += [0.5 + t*(0.3 + 3.9*t - 4.7*t**2)]
    y += [1.5 + t*(0.3 + 0.9*t - 2.7*t**2)]
    axTStar.plot(r, s, color='red')
    axTStar.plot(x, y, 'bo', color='blue')
    cameraTStar.snap()
animTStar = cameraTStar.animate(blit=False, interval=100)
animTStar.save('tStar.mp4')
```
