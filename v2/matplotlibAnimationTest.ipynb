```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation, rc

from IPython.display import HTML
```

```python
fig = plt.figure()
fig.set_dpi(100)
fig.set_size_inches(4, 4)

ax = plt.axes(xlim=(0, 2), ylim=(0, 2))
patch = plt.Circle((0.5, 1.5), 0.1, fc='y')
```

```python
def init():
    patch.center = (0.5, 1.5)
    ax.add_patch(patch)
    return patch,
```

```python
def animate(t):
    x, y = patch.center
    
    x = 0.5 + 0.001*t*(0.3 + 3.9*t - 4.7*t**2)
    y = 1.5 + 0.001*t*(0.3 + 0.9*t - 2.7*t**2)
    
    # x = 0.5 + 0.003*t 
    # y = 1.5 + 0.009*t
    
    patch.center = (x, y)
    return patch,
```

```python
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=None, interval=200, blit=True)
plt.show()
```
