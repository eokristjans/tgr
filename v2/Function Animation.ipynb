```python
# See: https://stackoverflow.com/questions/53435352/animate-matplotlib-parametric-example
import matplotlib
matplotlib.use('Agg')
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import animation, rc
import matplotlib.pyplot as plt
import math
from celluloid import Camera

import functionsV2
import v2
```

```python
plt.rcParams['legend.fontsize'] = 10
fig = plt.figure()
ax = fig.gca()
camera = Camera(fig)
x=[]
y=[]
r=[]
s=[]


# Prepare arrays r, s
for t in np.arange(0, 1, 0.01):
    print(round(t,2))
    r += [0.5 + t*(0.3 + 3.9*t - 4.7*t**2)]
    s += [1.5 + t*(0.3 + 0.9*t - 2.7*t**2)]

# Prepare arrays x, y
for t in v2.tStarArray:
    x += [0.5 + t*(0.3 + 3.9*t - 4.7*t**2)]
    y += [1.5 + t*(0.3 + 0.9*t - 2.7*t**2)]
    ax.plot(r, s, color='red')
    ax.plot(x, y, 'bo', color='blue')
    camera.snap()
anim = camera.animate(blit=False, interval=100)
#anim.save('2d.mp4')
```
