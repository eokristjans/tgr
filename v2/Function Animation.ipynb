```python
# See: https://stackoverflow.com/questions/53435352/animate-matplotlib-parametric-example
import matplotlib
matplotlib.use('Agg')
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import animation, rc
import matplotlib.pyplot as plt
import math
from celluloid import Camera

plt.rcParams['legend.fontsize'] = 10
fig = plt.figure()
ax = fig.gca()
camera = Camera(fig)
x=[]
y=[]
# Prepare arrays x, y
for t in np.arange(0, 1, 0.01):
    x += [0.5 + t*(0.3 + 3.9*t - 4.7*t**2)]
    y += [1.5 + t*(0.3 + 0.9*t - 2.7*t**2)]
    ax.plot(x, y, color='blue')
    camera.snap()
anim = camera.animate(blit=False, interval=10)
anim.save('2d.mp4')
```
